# Welcome to Colab!

```
import torch
```

```
if torch.cuda.is_available():
    device = torch.cuda.get_device_name(0)
else:
    device = torch.device("cpu")
print(device)
```

## Creating a Tensor

```
# Using empty()
torch.empty(2,3)
```

```
a= torch.empty(2,3)
print(type(a))
```

```
## Using zeroes
print(torch.zeros(2,3))
```

```
## Using ones
print(torch.ones(2,3))
```

```
## Using rand
print(torch.rand(2,3))
```

```
## Using seed
torch.manual_seed(42)   #It fixed the values
print(torch.rand(2,3))
```

```
## Using tensor
print(torch.tensor([[1,2,3],[4,5,6]]))
```

```
##Using arange
print(torch.arange(0,10))
```

## Shape of Tensor


```
x = torch.tensor([[1,2,3],[4,5,6]])
x
```

```
x.shape
```

```
print(torch.empty_like(x))   ## Copy the same tensor but with x-like config
```

```
print(torch.ones_like(x))
```

```
print(torch.rand_like(float(x)))
```

## Tensor Data Type


```
x.dtype
```

```
## assign data type
print(torch.tensor([[1.0,2.3,3.1],[4.3,5.4,6.3]] , dtype= torch.int32 ))
```

## Change to CUDA I.E. GPU


```
device = torch.device('cuda')
print(device)
```

```
x.to(device=device)           ## Copy the tensor array to CUDA
```

## Convert Tensor to Numpy


```
y=x.numpy()             #Conver tensor type into numpy
y
```

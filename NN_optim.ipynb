```
import torch
import torch.nn as nn
```

```
class Model(nn.Module):
  def __init__(self, num_features) -> None:
    super().__init__()   # calling the constructor of nn.Module class to use nn.Linear & nn.Sigmoid

    # Create Layer
    self.linear = nn.Linear(num_features, 1) #nn.Linear(input,output)
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features)  #Linear Transformation
    out = self.sigmoid(out)      #Activation Function
    return out


```

## NN using hidden layer


```
class Model(nn.Module):
  def __init__(self, num_features) -> None:
    super().__init__()   # calling the constructor of nn.Module class to use nn.Linear & nn.Sigmoid

    # Create Layers
    self.linear1 = nn.Linear(num_features, 3) #nn.Linear(input,output)
    self.relu = nn.ReLU()  # For hidden layer/s activation function is RELU
    self.linear2 = nn.Linear(3,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear1(features)  #Linear Layer 1 Transformation
    out = self.relu(out)          #Activation Function for hidden layer
    out = self.linear2(out)      #Linear Layer 2 Transformation
    out = self.sigmoid(out)      #Activation Function
    return out


```

```
#create  dataset
features = torch.rand(10,5)  # rows = 10 & cols = 5

#create model

model = Model(features.shape[1])  # take the cols as an arguemnt


# call Model fo forward pass
#model = model.forward(features)

# Due to magic function __call__ by default if you call an object of the class, NN class would call forward function

model(features)
```

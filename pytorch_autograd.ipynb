## Using Autograd

```
import torch
```

```
x= torch.tensor(3.0,requires_grad=True) # Requires grad means pytorch keeps track in order to compute the differentiation
```

```
x
```

```
y=x**2
y
```

```
## PyTorch makes the Computation graph for the above
# Like this   x--sqrt--y
# That's why it knows grad_fn=<PowBackward0> for y
```

```
y.backward()
```

```
x.grad  # To calculate the differentiation
```

```
a=torch.tensor(4.0,requires_grad=True)
a
```

```
b=a**2
b
```

```
c=torch.sin(b)
c
```

```
## Computation Graph created by the above

# x--> sqrt --> y --> sin --> c
```

```
c.backward()   # To calculate the  derivative
```

```
a.grad
```

```
import torch
x=torch.tensor(6.7)
y=torch.tensor(0.0)
w= torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```
w
```

```
b
```

```
#Forward Pass
z=w*x+b         # Linear Transformation
z
```

```
y_pred = torch.sigmoid(z)  # Activation function
y_pred
```

```
import torch.nn.functional as F

# Loss function
loss = F.binary_cross_entropy(y_pred, y)
loss
```

```
loss.backward()
```

```
print(w.grad)
print(b.grad)
```

```
x = torch.tensor([1.0,2.0,3.0],requires_grad=True)
x
```

```
y=(x**2).mean()
y
```

```
y.backward()
```

```
x.grad   # getting 3 gradients one for each input of x
```

```
# clearing grad
x.grad.zero_()
```
